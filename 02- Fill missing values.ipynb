```python
missing_value_per = df.isnull().sum()/df.shape[0] *100
missing_value_per
```

```python
missing_value_clm_gre_20 = missing_value_per[missing_value_per > 20].keys()
missing_value_clm_gre_20
```

```python
df2_drop_clm = df.drop(columns = missing_value_clm_gre_20)
df2_drop_clm.shape
```

```python
df3_num = df2_drop_clm.select_dtypes(include=['int64','float64'])
df3_num.head()
```

```python
# List comprehension to get column names with missing values
columns_with_missing_values = [
    var  # Column name
    for var in df3_num.columns  # Iterate over each column in df3_num
    if df3_num[var].isnull().sum() > 0  # Include only if there are missing values in the column
]

# Print the list of column names with missing values
print("Columns with Missing Values:")
print(columns_with_missing_values)
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(columns_with_missing_values):
    plt.subplot(2,2,i+1)
    sns.distplot(df3_num[var], bins=20, kde_kws={'linewidth':5, 'color':'#DC143C'})
    
```

```python
# mean
df4_num_mean = df3_num.fillna(df3_num.mean())
```

```python
df4_num_mean.isnull().sum().sum()
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(missing_num_var):
    plt.subplot(2,2,i+1)
    sns.distplot(df3_num[var], bins=20, kde_kws={'linewidth':8, 'color': 'red'}, label='Orignal')
    sns.distplot(df4_num_mean[var], bins=20, kde_kws={'linewidth':5, 'color': 'green'}, label='Mean')
    
```

```python
# median
df5_num_median = df3_num.fillna(df3_num.median())
df5_num_median.isnull().sum().sum()
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(missing_num_var):
    plt.subplot(2,2,i+1)
    sns.distplot(df3_num[var], bins=20,hist=False, kde_kws={'linewidth':8, 'color': 'red'}, label='Orignal')
    sns.distplot(df4_num_mean[var], bins=20,hist=False, kde_kws={'linewidth':5, 'color': 'green'}, label='Mean')
    sns.distplot(df5_num_median[var], bins=20,hist=False, kde_kws={'linewidth':3, 'color': 'yellow'}, label='Median')
plt.legend()    
```

```python
for i, var in enumerate(missing_num_var):
    plt.figure(figsize=(10,10))
    plt.subplot(3,1,1)
    sns.boxplot(df[var])
    plt.subplot(3,1,2)
    sns.boxplot(df4_num_mean[var])
    plt.subplot(3,1,3)
    sns.boxplot(df5_num_median[var])
```

```python
df_concat = pd.concat([df3_num[missing_num_var], 
                       df4_num_mean[missing_num_var],
                       df5_num_median[missing_num_var]],axis=1 )
```
